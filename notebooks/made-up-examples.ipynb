```python
import matplotlib.pyplot as plt 
import numpy as np

from hgmc.plots import plot_distribution
```

```python
very_small_data = np.array([
    [1, 3, 9],
    [0, 8, 9],
    [4, 5, 6],
    [0, 2, 4],
    [5, 6, 8],
    [2, 4, 5],
])
num_bins = very_small_data.max() + 1
```

```python
clusters_level_0 = very_small_data
coverage_level_0 = np.histogram(clusters_level_0, bins=np.arange(num_bins + 1))[0]

max_coverage = coverage_level_0.max()
yticks = np.arange(max_coverage + 1)

plot_distribution(
    coverage_level_0,
    colors='#999999',
    title='Coverage by all clusters',
    figsize=(13, 1.25),
    yticks=yticks
)

clusters_level_1 = very_small_data[(very_small_data == 4).sum(axis=1).astype(bool)]
coverage_level_1 = np.histogram(clusters_level_1, bins=np.arange(num_bins + 1))[0]

plot_distribution(
    coverage_level_1,
    colors=['#999999'] * 4 + ['#2854c5'] + ['#999999'] * 4,
    title='Coverage by clusters anchoring at bin #4',
    figsize=(13, 1.25),
    yticks=yticks
)

clusters_level_2 = clusters_level_1[(clusters_level_1 == 5).sum(axis=1).astype(bool)]
coverage_level_2 = np.histogram(clusters_level_2, bins=np.arange(num_bins + 1))[0]

plot_distribution(
    coverage_level_2,
    colors=['#999999'] * 4 + ['#2854c5'] + ['#6350a2'] + ['#999999'] * 3,
    title='Coverage by clusters anchoring at bin #4 and #5',
    figsize=(13, 1.25),
    yticks=yticks,
    ylim=(0, max_coverage),
)
```

```python
from hgmc.plots import plot_distribution

bold_colors = ['#990200', '#38761e', '#b45f06', '#c09209', '#999999']
lite_colors = ['#dd7e6b', '#93c47d', '#f6b26b', '#fbd967', '#d9d9d9']
labels = ['P', 'T', 'AE', 'PE', 'H']

# Query results
Y_1_1 = [2, 2, 1, 0, 0] # All unique covered bins from cluster 4 and 5
Y_1_2 = [2, 2, 2, 0, 0] # All bins from clusters 4 and 5

Y_2_1 = [3, 2, 1, 1, 3] # All unique bins
Y_2_2 = [3, 2, 1, 1, 2] # All unique covered bins from all clusters
Y_2_3 = [6, 5, 3, 2, 2] # All bins from all clusters
Y_2_4 = [5, 5, 3, 1, 0] # All bins from clusters that ove|rlap with clusters 4 and 5 (incl. cluster 4 and 5)
Y_2_5 = [3, 3, 1, 1, 0] # All bins from clusters that overlap with clusters 4 and 5 (excl. cluster 4 and 5)

plot_distribution(Y_1_2, labels, Y_comp=Y_2_4, colors=lite_colors, figsize=(6, 4))
```

## Little bit bigger example

```python
small_data = np.array([
    [4, 8, 10, np.nan],
    [1, 3, 9, np.nan],
    [0, 8, 9, np.nan],
    [7, 8, 9, 11],
    [4, 5, 6, np.nan],
    [0, 2, 4, np.nan],
    [5, 6, 8, np.nan],
    [2, 4, 5, np.nan],
])
num_bins = np.nanmax(small_data).astype(int) + 1
```

```python
clusters_level_0 = small_data
coverage_level_0 = np.histogram(clusters_level_0, bins=np.arange(num_bins + 1))[0]

max_coverage = coverage_level_0.max()
yticks = np.arange(max_coverage + 1)

plot_distribution(
    coverage_level_0,
    colors='#999999',
    title='Coverage by all clusters',
    figsize=(13, 1.25),
    yticks=yticks
)

clusters_level_1 = very_small_data[(very_small_data == 4).sum(axis=1).astype(bool)]
coverage_level_1 = np.histogram(clusters_level_1, bins=np.arange(num_bins + 1))[0]

plot_distribution(
    coverage_level_1,
    colors=['#999999'] * 4 + ['#2854c5'] + ['#999999'] * 4,
    title='Coverage by clusters anchoring at bin #4',
    figsize=(13, 1.25),
    yticks=yticks
)

clusters_level_2 = clusters_level_1[(clusters_level_1 == 5).sum(axis=1).astype(bool)]
coverage_level_2 = np.histogram(clusters_level_2, bins=np.arange(num_bins + 1))[0]

plot_distribution(
    coverage_level_2,
    colors=['#999999'] * 4 + ['#2854c5'] + ['#6350a2'] + ['#999999'] * 3,
    title='Coverage by clusters anchoring at bin #4 and #5',
    figsize=(13, 1.25),
    yticks=yticks,
    ylim=(0, max_coverage),
)
```

```python
from hgmc.plots import plot_distribution

bold_colors = ['#990200', '#38761e', '#b45f06', '#c09209', '#999999']
lite_colors = ['#dd7e6b', '#93c47d', '#f6b26b', '#fbd967', '#d9d9d9']
labels = ['P', 'T', 'AE', 'PE', 'H']

# Query results
Y_1_1 = [2, 2, 2, 0, 0] # All unique bins covered by clusters 0, 4, and 5
Y_1_2 = [3, 3, 3, 0, 0] # All bins from clusters 0, 4, and 5

Y_2_0 = [3, 2, 2, 1, 4] # All unique bins from all clusters
Y_2_1a = [3, 2, 1, 1, 2] # All unique bins from clusters 2, 3, 4, and 5
Y_2_1b = [3, 2, 1, 1, 4] # All unique bins from clusters 1, 2, 3, 4, and 5
Y_2_2a = [5, 3, 1, 2, 2] # All bins from clusters 2, 3, 4, and 5
Y_2_2b = [5, 3, 1, 3, 4] # All bins from clusters 1, 2, 3, 4, and 5

plot_distribution(
    Y_1_1,
    labels,
    Y_comp=Y_2_1a,
    colors=lite_colors,
    figsize=(7, 3.5),
    title='Absolute Feature Counts by Unique Cluster Bins',
    primary_label='Bins from clusters 0,6,7',
    comparison_label='Bins from clusters 2,3,4,5'
)
plot_distribution(
    Y_1_1,
    labels,
    Y_comp=Y_2_1b,
    colors=lite_colors,
    figsize=(7, 3.5),
    title='Absolute Feature Counts by Unique Cluster Bins',
    primary_label='Bins from clusters 0,6,7',
    comparison_label='Bins from clusters 1,2,3,4,5'
)
plot_distribution(
    Y_1_2,
    labels,
    Y_comp=Y_2_2a,
    colors=lite_colors,
    figsize=(7, 3.5),
    title='Absolute Feature Counts by Cluster Bins',
    primary_label='Bins from clusters 0,6,7',
    comparison_label='Bins from clusters 2,3,4,5'
)
plot_distribution(
    Y_1_2,
    labels,
    Y_comp=Y_2_2b,
    colors=lite_colors,
    figsize=(7, 3.5),
    title='Absolute Feature Counts by Cluster Bins',
    primary_label='Bins from clusters 0,6,7',
    comparison_label='Bins from clusters 1,2,3,4,5'
)
```
